```python
from autoscraper import AutoScraper
import pandas as pd
```

```python
url="https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"

wanted_list=["₹34,999","APPLE iPhone SE (Red, 128 GB)","128 GB ROM","11.94 cm (4.7 inch) Retina HD Display","12MP Rear Camera | 7MP Front Camera","A13 Bionic Chip with 3rd Gen Neural Engine Processor"]
```

```python
scraper=AutoScraper()
result=scraper.build(url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_i4pd':'Title','rule_75gl':'Price','rule_p11s':'ROM','rule_f2l9':'Display','rule_04y5':'Camera','rule_kaas':'Processor'})
scraper.keep_rules(['rule_i4pd','rule_75gl','rule_p11s','rule_f2l9','rule_04y5','rule_kaas'])
scraper.save('flipkart-search')
```

```python
#Just add new Flipkart URL for different results.
results=scraper.get_result_similar('https://www.flipkart.com/search?q=samsung+mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_na&as-pos=1&as-type=RECENT&suggestionId=samsung+mobiles%7CMobiles&requestId=a2b58de8-99c1-406a-8a9b-d4fc7ed0e06b&as-searchtext=Samsung&page=2',group_by_alias=True)
```

```python
#Program to extract data of all pages into single csv in single program
```

```python
urlscrape='https://www.flipkart.com/search?q=samsung+mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_na&as-pos=1&as-type=RECENT&suggestionId=samsung+mobiles%7CMobiles&requestId=a2b58de8-99c1-406a-8a9b-d4fc7ed0e06b&as-searchtext=Samsung&p%5B%5D=facets.price_range.from%3D4000&p%5B%5D=facets.price_range.to%3DMax&sort=popularity'
```

```python
df3 = pd.DataFrame(columns =['Name', 'Price','Storage','Battery','Camera','Display'])                   
for i in range(1,28):
    urlscraper=urlscrape+'&page='+str(i)
    results=scraper.get_result_similar(urlscraper,group_by_alias=True)
    Camera=results['Camera']
    Price=results['Price']    
    Name=results['Title']
    Storage=results['ROM']
    Display=results['Display']
    Processor=results['Processor']
    df2 = pd.DataFrame(list(zip(Name,Price,Storage,Processor,Camera,Display)),
               columns =['Name', 'Price','Storage','Battery','Camera','Display'])
    df3 = df3.append(df2)
```

```python
df3
```

```python
df3.to_csv('Flipkart Samsung Full Scrap.csv',index=False,header=True)
```
